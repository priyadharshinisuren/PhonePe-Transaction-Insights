```python
path = "C:/Users/priya/phonepe/pulse/data/aggregated/user/country/india/state/"
user_state = os.listdir(path)

df3 = {
    'State': [], 
    'Year': [], 
    'Quarter': [], 
    'brand': [],
    'appOpens': [], 
    'registeredUsers': [], 
    'transaction_percentage': []
}

for i in user_state:
    p_i = os.path.join(path, i)
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        p_j = os.path.join(p_i, j)
        Agg_yr_list = os.listdir(p_j)
        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)
            with open(p_k, 'r') as Data:
                D = json.load(Data)

            if D.get('data') and D['data'].get('usersByDevice'):
                for z in D['data']['usersByDevice']:
                    df3['appOpens'].append(z.get('appOpens', 0))
                    df3['brand'].append(z.get('brand', 'Unknown'))
                    df3['registeredUsers'].append(z.get('count', 0))
                    df3['transaction_percentage'].append(z.get('percentage', 0))
                    df3['State'].append(i)
                    df3['Year'].append(j)
                    df3['Quarter'].append(int(k.replace('.json', '')))

Agg_user = pd.DataFrame(df3)
```

```python
path = "C:/Users/priya/phonepe/pulse/data/map/insurance/hover/country/india/state/"
map_state = os.listdir(path)

df1 = {
    'State': [], 
    'Year': [], 
    'Quarter': [], 
    'District': [],
    'insurance_count': [],
    'insurance_amount': []
}

for i in map_state:
    p_i = os.path.join(path, i)
    map_yr = os.listdir(p_i)
    
    for j in map_yr:
        p_j = os.path.join(p_i, j)
        Agg_yr_list = os.listdir(p_j)
        
        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)
            
            with open(p_k, 'r') as Data:
             D = json.load(Data)

# Case 1: district-level transactions
            if "districts" in D["data"]:
                 for z in D["data"]["districts"]:
                    if "metric" in z:  # transaction schema
                     name = z["entityName"] if "entityName" in z else z["name"]
                     count = z["metric"].get("count", 0)
                     amount = z["metric"].get("amount", 0)
                    df1["District"].append(name)
                 df1["insurance_count"].append(count)
                 df1["insurance_amount"].append(amount)
                 df1["State"].append(i)
                 df1["Year"].append(j)
                 df1["Quarter"].append(int(k.strip(".json")))

# Case 2: hoverDataList schema
            elif "hoverDataList" in D["data"]:
                for z in D["data"]["hoverDataList"]:
                  name = z["name"]
                  metric = z["metric"][0]
                  count = metric.get("count", 0)
                  amount = metric.get("amount", 0)
                df1["District"].append(name)
                df1["insurance_count"].append(count)
                df1["insurance_amount"].append(amount)
                df1["State"].append(i)
                df1["Year"].append(j)
                df1["Quarter"].append(int(k.strip(".json")))

map_insurance=pd.DataFrame(df1)
```

```python
path = "C:/Users/priya/phonepe/pulse/data/map/user/hover/country/india/state/"
map_state = os.listdir(path)

df3 = {
    'State': [], 
    'Year': [], 
    'Quarter': [], 
    'District': [], 
    'appOpens': [], 
    'registeredUsers': [] 
}

for i in map_state:  
    p_i = path + i + "/"  
    map_yr = os.listdir(p_i)  
    for j in map_yr:  
        p_j = p_i + j + "/"  
        map_yr_list = os.listdir(p_j)  
        for k in map_yr_list:  
            p_k = p_j + k  
            with open(p_k, 'r') as Data:  
                D = json.load(Data)

            # Extract district-level hoverData
            if 'hoverData' in D['data'] and D['data']['hoverData'] is not None:
                for district, values in D['data']['hoverData'].items():
                    df3['District'].append(district)
                    df3['registeredUsers'].append(values['registeredUsers'])
                    df3['appOpens'].append(values['appOpens'])
                    df3['State'].append(i)
                    df3['Year'].append(j)
                    df3['Quarter'].append(int(k.strip('.json')))

map_user = pd.DataFrame(df3)
```

```python
import os
import json
import pandas as pd

path = "C:/Users/priya/phonepe/pulse/data/top/user/country/india/state/"
top_state = os.listdir(path)

df = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'District': [],
    'Pincode': [],
    'RegisteredUsers': []
}

for i in top_state:  # loop through states
    p_i = os.path.join(path, i)
    top_yr = os.listdir(p_i)

    for j in top_yr:  # loop through years
        p_j = os.path.join(p_i, j)
        top_yr_list = os.listdir(p_j)

        for k in top_yr_list:  # loop through quarters
            p_k = os.path.join(p_j, k)

            with open(p_k, 'r') as Data:
                D = json.load(Data)

            # ✅ Extract district-level user data
            if "districts" in D["data"]:
                for item in D["data"]["districts"]:
                    df['District'].append(item['name'])
                    df['Pincode'].append(None)  # no pincode in districts
                    df['RegisteredUsers'].append(item['registeredUsers'])
                    df['State'].append(i)
                    df['Year'].append(j)
                    df['Quarter'].append(int(k.strip('.json')))

            # ✅ Extract pincode-level user data
            if "pincodes" in D["data"]:
                for item in D["data"]["pincodes"]:
                    df['District'].append(None)  # no district in pincodes
                    df['Pincode'].append(item['name'])
                    df['RegisteredUsers'].append(item['registeredUsers'])
                    df['State'].append(i)
                    df['Year'].append(j)
                    df['Quarter'].append(int(k.strip('.json')))

# Convert to DataFrame
top_user = pd.DataFrame(df)
```
